1. Define a function named “triple_and” that takes three parameters and returns True
only if they are all True and False otherwise

```python
def triple_and(a,b,c):
    if (a==b==c==True):
        return True
    else:
        return False
print(triple_and(10==10, 10==10, 10==10))
print(triple_and(10==10, 10==10, 10==100))
```

2. Write a python program to sort a string alphabetically and print the count of each
character

```python
def sorter(string):
    string.sort()
    print(string)
    return string

def count(string):
    unique_string = list(set(string))
    for i in range(len(unique_string)):
        print(f"Count of {unique_string[i]} is {string.count(unique_string[i])}")

count(sorter(list(input("Enter the string: "))))
```

3. Write a python program to check if given number is a hill number.

```python
# A Hill number is one which goes from low to high then high to low so it goes like a hill.... 

# The digits strictly increase from the first digit up to the largest digit. 
# After reaching the largest digit, the digits strictly decrease toward the last digit. 
# The largest digit can be repeated, but only consecutively 
# (no smaller digits can appear between repetitions). 

def hill_number(n):
    uphill = True #Flag variable
    
    for i in range(0, len(n)-1):
        if int(n[i]) < int(n[i+1]) and uphill==True:        # 1 2 3 4 7 7 7 3 1
            uphill = True
            print("uphill")
        elif int(n[i]) == int(n[i+1]):
            uphill = False
            print("constant")
        elif int(n[i]) > int(n[i+1]) and uphill==False:
            uphill = False
            print("downhill")
        else:
            return "No, It is not a hill number"
    return "Yes, it is a hill number"
print(hill_number(input("Enter the number: ")))         # 123477731
```

4. Write a python function to perform selection sort on a given string.

```python
def selection_sort(string):
    string = list(string)
    length = len(string)

    for i in range(length):
        min = i
        for j in range(i, length):
            if string[min] > string[j]:
                min = j
        tmp = string[i]
        string[i] = string[min]
        string[min] = tmp

    return string
print(selection_sort(input("Enter the string: ")))
```

5. Find the factorial of a given number using recursion.


```python
def factorial(num):
    if num==1 or num==0:
        return 1
    else:
        return num*(num-1)
print("The factorial is: ", factorial(input("Enter the number: ")))
```

6. Create a function that checks whether given string is an anagram or not?


```python
def anagram_check(string):
    if len(list(string)) == len(list(set(string))):
        return "Yes, it is an anagram"
    else:
        return "No, it is not an anagram"
print(anagram_check(input("Enter the number: ")))
```

7. Write a program to print the Fibonacci Sequence till n-values where n is user input.

```python
def fibonacci(a, b, n):
    if n==0:
        return a
    else:
        n-=1
        print(a, end=" ")
        return fibonacci(b, a+b, n)
print(fibonacci(0, 1, int(input("Enter number: "))))
```

8. Write a python program to divide a given string into equal parts containing n (user
input) characters of same sequence.

Example:    string=“abcdabcdabcdabcd”

n=4

output: “abcd”, “abcd”, “abcd”, “abcd”

If the division is not possible or the sequence cannot be same, print out the
appropriate error.


```python
string = input("Enter the squence: ")
n = int(input("Enter number: "))
if len(string)%n != 0:
    raise ValueError("Error occured as string can't be split perfectly")
else:
    seq = string[0:n]
    for i in range(0, len(string), n):
        if seq != string[i:i+n]:
            raise ValueError("No equal parts of string")
        else:
            print(string[i:i+n])
```
