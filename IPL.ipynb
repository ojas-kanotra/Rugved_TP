Q10. Find the players who have won ‘Player of the Match’ more than 3 times.

```python
pom = dict(df['player_of_match'].value_counts())
for i in list(pom.keys()):
    if pom[i] > 3:
        print(f"{i}\t{pom[i]}")
```
